## Printing Missing Values

```python
def check_null_values(df):
    null_values = df.isnull()
    
    for row in range(null_values.shape[0]):
        for col in range(null_values.shape[1]):
            if null_values.iloc[row, col]:
                print(f'Missing value at row index: {row}, column index: {col}')
```

```python
check_null_values(df)
```
